```python
### Importing The Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Reading the File
df = pd.read_csv('/content/data.csv',encoding='latin1')
df.head()
```

```python
### Taking Our Required Column
x=df['no2']
x.shape
```

```python
### Checking for NULL values
x.isnull().sum()
```

```python
### Droping the NULL values
x.dropna(inplace=True)
x.shape
```

```python
### Defining My Roll Number
R=102303806
```

```python
### Computing the Transformation cofficients
ar = 0.05*(R % 7)
br = 0.3 *((R % 5) + 1)
print(f'ar coffiecient = {ar}')
print(f'br coffiecient = {br}')
```

```python
### The Transformation of X into Z
z = x + ar * np.sin(br * x)
z.shape
```

```python
### checking the z head
z.head()
```

```python
### The Equation given in the question is the eqation of the Gaussian Dist
### The form p(z)=1/sqrt(2*pi*var) * e^((z-u)^2/(2*var))
```

```python
### Calculating the Values of lam,u,c
u = np.mean(z)
var = np.var(z,ddof=0)
lam = 1/(2*var)
c=1/np.sqrt(2*np.pi*var)
```

```python
### Printing the Values
print(f'u value : {u}')
print(f'lam value : {lam}')
print(f'c value : {c}')
```

```python
### Rounding the values
u=round(u,6)
lam=round(lam,6)
c=round(c,6)
```

```python
### Printing The Rounded Values
print(f'u value : {u}')
print(f'lam value : {lam}')
print(f'c value : {c}')
```
